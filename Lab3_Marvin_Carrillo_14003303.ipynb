1. ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
import array

# Creación un array con 1,000 elementos de tipo booleano
arr = array.array('b', [False] * 1000)

# Cálculo de tamaño en bytes del array
size_bytes = arr.itemsize * len(arr)

print("El tamaño en bytes de un array de 1,000 elementos de tipo booleano es:", size_bytes)
```

3. Dado el array:

>2 4 5 6

>0 3 7 4

>8 8 5 2

>1 5 6 1

>Seleccione con una instrucción el subarray de elementos
>0 3 7 4

>Después, seleccione el subarray de elementos
>2 5
>8 5

```python
import numpy as np

# Creación de la matriz brindada
matriz = np.array([[2, 4, 5, 6],
                    [0, 3, 7, 4],
                    [8, 8, 5, 2],
                    [1, 5, 6, 1]])

# Seleccionar el subarray de elementos 0 3 7 4
subarray1 = matriz[1, :]
print("Subarray de elementos 0 3 7 4:")
print(subarray1)

# Seleccionar el subarray de elementos 2 5
subarray2 = matriz[[0,2], [0,2]]
print("\nSubarray de elementos 2 5:")
print(subarray2)

# Seleccionar el subarray de elementos 8 5
subarray3 = matriz[[2,2], [0,2]]
print("\nSubarray de elementos 8 5:")
print(subarray3)
```

4. Dados los arrays 

> a = [1, 4, 2, 7] y 

> b = [1, 3, 2, 9]

> obtenga la media aritmética de la diferencia (a-b).

```python
import numpy as np

# Definición de los arrays a y b
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calcular la diferencia entre los arrays a y b
diferencia = a - b

# Calcular la media aritmética de las diferencias
media_diferencia = np.mean(diferencia)

print("La media aritmética de la diferencia entre a y b es:", media_diferencia)
```

5. Tengo valores de coordenadas (x, y) en las columnas del array:

>1.33 4.5
>30.0 10.7
>70.2 0.5

>Agregue a este array las coordenadas (37.1, -3.6). 

> Muestre en pantalla las dimensiones del nuevo array.

```python
import numpy as np

# Definición del array inicial con las coordenadas existentes
coordenadas = np.array ([[1.33, 4.5],
                         [30.0, 10.7],
                         [70.2, 0.5]])

# Adición de las coordenadas (37.1, -3.6) al array existente
nuevas_coordenadas = np.array([[37.1, -3.6]])
coordenadas = np.vstack([coordenadas, nuevas_coordenadas])

# Mostrar las dimensiones del nuevo array
print("Dimensiones del nuevo array de coordenadas:")
print(coordenadas.shape)
```

```python
import numpy as np

# Definición del array inicial con las coordenadas existentes
coordenadas = np.array([[1.33, 4.5],
                         [30.0, 10.7],
                         [70.2, 0.5]])

# Trasponer el array de coordenadas
coordenadas_traspuesta = coordenadas.T

# Adición de 2 nuevos pares de coordenadas al array traspuesto
nuevas_coordenadas = np.array([[10.8, 3.0],
                               [35.8, 12.0]])
coordenadas_traspuesta = np.hstack([coordenadas_traspuesta, nuevas_coordenadas.T])

# Mostrar las dimensiones del nuevo array traspuesto
print("Dimensiones del nuevo array traspuesto de coordenadas:")
print(coordenadas_traspuesta.shape)
```

7. Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan.

```python
# Selección al azar cincuenta municipios diferentes
municipios_aleatorios = datasetlab3['Municipio'].sample(n=50, replace=False)

# Filtro del dataset solo con los municipios seleccionados
municipios_seleccionados = datasetlab3[datasetlab3['Municipio'].isin(municipios_aleatorios)]

# Cálculo del promedio y la desviación estándar de la población de los cincuenta municipios
promedio_poblacion = municipios_seleccionados['Poblacion'].mean()
desviacion_estandar_poblacion = municipios_seleccionados['Poblacion'].std()

print("Promedio de la población de los cincuenta municipios seleccionados:", promedio_poblacion)
print("Desviación estándar de la población de los cincuenta municipios seleccionados:", desviacion_estandar_poblacion)
```

8. Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python

# Encontrar las coordenadas de Madrid
coordenadas_madrid = datasetlab3.loc[datasetlab3['Municipio'] == 'Madrid', ['X', 'Y']]

# Extraer las coordenadas de Madrid
x_madrid = coordenadas_madrid['X'].values[0]
y_madrid = coordenadas_madrid['Y'].values[0]

print("Las coordenadas de Madrid son:")
print("Coordenada X:", x_madrid)
print("Coordenada Y:", y_madrid)

# Coordenadas de Madrid
coord_madrid = (x_madrid, y_madrid)

# Calcular la distancia entre cada municipio y Madrid
datasetlab3['Distancia_Madrid'] = datasetlab3.apply(lambda row: np.sqrt((row['X'] - x_madrid)**2 + (row['Y'] - y_madrid)**2), axis=1)

# Excluir Madrid y encontrar el municipio más cercano y más lejano
municipio_mas_cercano = datasetlab3.loc[datasetlab3['Distancia_Madrid'].idxmin()]['Municipio']
municipio_mas_lejano = datasetlab3.loc[datasetlab3['Distancia_Madrid'].idxmax()]['Municipio']

print("El municipio más cercano a Madrid es:", municipio_mas_cercano)
print("El municipio más lejano a Madrid es:", municipio_mas_lejano)
```
