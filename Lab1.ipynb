# **¿Qué es Git?**

> Git es un sistema de control de versiones distribuido, diseñado para manejar proyectos de software de forma eficiente. Permite a los equipos trabajar en el mismo código fuente, realizar un seguimiento de los cambios y colaborar de forma eficaz.


> ### **Flujo de trabajo basado en Git**

> * Crear una rama a partir de la copia principal de los archivos en los que el usuario y los colaboradores están trabajando.
> * Realizar ediciones en los archivos de forma independiente y segura en la propia rama personal.
> * Git combina de manera inteligente los cambios específicos en la copia principal de los archivos, para que los cambios no afecten las actualizaciones de otras personas.
> * Git permite dar un seguimiento de los cambios del usuario y los de otras personas, para que todos sigan trabajando en la versión más actualizada del proyecto.



## **Puntos y características Clave**

> ### **Puntos Clave**

> - Git es una herramienta esencial para el desarrollo de software.
> - Es fácil de aprender y usar.
> - Permite un desarrollo eficiente y colaborativo.
> - Es una herramienta muy utilizada en la industria del software.


> ### **Características Clave**

> - ***Control de versiones:*** Permite realizar un seguimiento de las modificaciones del código fuente a lo largo del tiempo.
> - ***Ramificación:*** Permite crear diferentes versiones del código fuente para trabajar en paralelo.
> - ***Colaboración:*** Facilita el trabajo en equipo, permitiendo fusionar cambios de diferentes colaboradores.
> - ***Distribuido:*** Cada desarrollador tiene una copia completa del repositorio, lo que facilita la colaboración sin conexión.


> ### **Terminología**

> - ***Repositorio:*** Es la colección de archivos que se gestiona con Git / Github.
> - ***Commit:*** Es una instantánea del estado del proyecto en un momento determinado.
> - ***Branch:*** Es una copia del repositorio que permite trabajar en una versión independiente del código.
> - ***Merge:*** Es el proceso de combinar dos o más ramas en una sola.
> - ***Pull request:*** Es una solicitud para que se fusionen los cambios de una rama en otra.








# **¿Qué es GitHub?**

> GitHub es una plataforma de alojamiento de código basada en Git. Permite a los usuarios compartir sus proyectos con el mundo, colaborar con otros desarrolladores y utilizar herramientas para la gestión del código.



## **Puntos y características Clave**

> ### **Puntos Clave**

> - GitHub es la plataforma de alojamiento de código más popular del mundo.
> - Es una herramienta gratuita para proyectos de código abierto.
> - Ofrece planes de pago para proyectos privados.
> - Es una herramienta muy utilizada por la comunidad de desarrolladores.


> ### **Características Clave**

> - ***Alojamiento de código:*** Permite alojar proyectos de forma gratuita o de pago.
> - ***Control de versiones:*** Permite realizar un seguimiento de las modificaciones del código fuente a lo largo del tiempo.
> - ***Colaboración:*** Facilita el trabajo en equipo, permitiendo crear y gestionar pull requests.
> - ***Herramientas:*** Ofrece herramientas para la gestión del código, como la revisión de código, la automatización de tareas y la gestión de tickets.


> ### **Terminología**

> - ***Repositorio:*** Es la colección de archivos que se gestiona con Git / Github.
> - ***Commit:*** Es una instantánea del estado del proyecto en un momento determinado.
> - ***Branch:*** Es una copia del repositorio que permite trabajar en una versión independiente del código.
> - ***Pull request:*** Es una solicitud para que se fusionen los cambios de una rama en otra.
> - ***Fork:*** Es una copia de un repositorio que se puede modificar y compartir de forma independiente.








# **¿Qué es MagicCells?**

> MagicCells es una extensión para Google Sheets que permite crear interfaces de usuario interactivas y personalizadas. Permite agregar gráficos, tablas, filtros y otros elementos para facilitar la visualización y el análisis de datos.



## **Puntos y características Clave**

> ### **Puntos Clave**

> - MagicCells es una herramienta poderosa para mejorar la visualización y el análisis de datos en Google Sheets.
> - Es fácil de aprender y usar.
> - Permite crear interfaces de usuario interactivas y personalizadas.
> - Es una herramienta muy útil para profesionales que trabajan con datos.


> ### **Características Clave**

> - ***Interfaz de usuario:*** Permite crear interfaces de usuario interactivas y personalizadas.
> - ***Visualización de datos:*** Permite crear gráficos, tablas y otros elementos para visualizar los datos de forma atractiva.
> - ***Análisis de datos:*** Permite agregar filtros, cálculos y otras funciones para analizar los datos.
> - ***Colaboración:*** Permite compartir las hojas de cálculo con otros usuarios y trabajar en equipo.


> ### **Terminología**

> - ***MagicCell:*** Es una celda que contiene un código JavaScript que permite crear elementos interactivos.
> - ***Dashboard:*** Es una vista personalizada que combina diferentes MagicCells para visualizar y analizar datos.
> - ***Widget:*** Es un elemento interactivo que se puede agregar a una hoja de cálculo, como un gráfico, una tabla o un filtro.
